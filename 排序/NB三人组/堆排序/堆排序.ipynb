# 堆 啥是堆？

堆是一种完全二叉树 结构特殊罢了

## 分为

### 大根堆

一颗完全二叉树，满足任一节点都比其孩子节点大

### 小根堆

一颗完全二叉树，满足任一节点都比其孩子节点小

# 堆的向下调整

假设：节点左右子树都是堆，单自身不是堆

为了保证堆的正确性(保证是小根堆或者大根堆的其中一组)会自行调整，小根堆小的上去 大根堆大的上去 浅显理解就是这样 或许我理解错了 请读者批评指正！

# 堆排序

堆排序利用了堆的向下调整的性质(不明白可以自行百度)

农村包围城市 从最小的子树开始

1.建立堆

2.得到堆顶元素，为最大元素

3.去掉堆顶，将堆的最后一个元素放置到堆顶，此时可以通过一次调整使堆先变的合理

4.堆顶元素始终为第二大元素(针对去掉的前一个元素来说)

5.重复步骤三，直到堆变空

## 应该咋写 堆排序

```python
def sift(lit: list, low: int, high: int) -> list:
    # 向下调整模块
    '''
    lit -> 列表
    high -> 堆的最后一个元素的位置
    low -> 堆顶
    '''
    i = low  # i最开始指向根节点
    j = 2 * i + 1  # j开始是左孩子
    tmp = lit[low]  # 把堆顶存储起来
    while j <= high:  # 只要j位置存在 就一直循环
        if lit[j + 1] > lit[j] and j + 1 <= high:  # 如果右孩子存在且比较大
            j = j + 1  # j指向右孩子
        if lit[j] > tmp:
            lit[i] = lit[j]
            i = j  # 往下看一层
            j = 2 * i + 1
        else:  # tmp更大些，把tmp放到i的位置为止
            lit[i] = tmp
            break
    lit[i] = tmp

    return lit
```
